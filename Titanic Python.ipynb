```python
#To know in which features there are NaN values
print(training.isnull().any())
print( )
print(test.isnull().any())
```

```python
#To show the missing value columns in tabular format we use the pandas package
def NaN_table(training, test):
    print("training data frame")
    print(pd.isnull(training).sum())
    print( )
    print("test data frame")
    print(pd.isnull(test).sum())
    
NaN_table(training, test)
```

```python
#Can remove the Cabin feature as it is having lot of missing values and at the same time can remove the Ticket feature
training.drop(labels = ["Cabin", "Ticket"], axis=1, inplace=True)
test.drop(labels=["Cabin", "Ticket"], axis=1, inplace=True)

NaN_table(training,test)
```

```python
training.info()
```

```python
training_clean = training[np.isfinite(training["Age"])]
training_clean.info()
```

```python
#Just to check the distribution of Age we had dropped the missing values above from the Age feature
plt.figure(figsize=(20,10))
plt.tight_layout()
#sns.countplot(training_clean["Age"])
sns.distplot(training_clean["Age"])
plt.title("Distribution of Age in training")
```

```python
test_clean = test[np.isfinite(test['Age'])]
```

```python
plt.figure(figsize=(20,10))
plt.tight_layout()
sns.distplot(test_clean['Age'])
plt.title("Distribution of Age in test")
```

```python
#From the above distribution we see that the graph is skewed right so we need to replace the missing values in Age with
#median values in the original dataset
training['Age'].fillna(training['Age'].median(), inplace=True)
test['Age'].fillna(test['Age'].median(), inplace=True)
training['Embarked'].fillna('S', inplace=True) 
test['Fare'].fillna(test['Fare'].median(), inplace=True)
```

```python
print(training.isnull().any())
print( )
print(test.isnull().any())
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
sns.barplot(x='Sex', y='Survived',data=training)
plt.title("The total male and female on the ship (survived and not survived included)")
plt.subplot(1,2,2)
sns.stripplot(x='Sex', y='Age', data=training)
plt.title("Distribution of male and female on the ship")

# It can be seen from below that the people (male/female) in the younger gap had a better chance of survival than the 
# older group.
```

```python
total_male_train_survived = training[training['Sex']=='male']['Survived'].sum()
total_female_train_survived = training[training['Sex']=='female']['Survived'].sum()

#The actual total survived people are (male + female)
print("The actual number of people who survived from the total people on the ship are: " + str(total_male_train_survived + total_female_train_survived))

#To see what proportion of total survived is male / female
print("Proportion of those survived is male: " + str(total_male_train_survived / (total_male_train_survived + total_female_train_survived)))
print("Proportion of those survived is female: " + str(total_female_train_survived / (total_female_train_survived+total_male_train_survived)))
```

```python
# The above graph shows that female population was more in number than male population on the ship as it is plotted based on Survived parameter which has 0 and 1 so count of records
# The proportion calculation shows that the percentage of those who survived were female more than male as it makes sense
# because women and children were evacuated first and then men were evacuated
```

```python
# Now lets see the same survival distribution based on class
plt.figure(figsize=(15,10))
plt.tight_layout()
sns.barplot('Pclass', 'Survived', data=training)
plt.title("The total passengers on the ship based on the class they were travelling (survived and not survived included)")
```

```python
# Total Passengers in each class
first = training['PassengerId'][training['Pclass']==1].count()
second = training['PassengerId'][training['Pclass']==2].count()
third = training['PassengerId'][training['Pclass']==3].count()

# Now lets find the total survivers per class
total_first = training[training['Pclass']==1]['Survived'].sum() 
total_second = training[training['Pclass']==2]['Survived'].sum() 
total_third = training[training['Pclass']==3]['Survived'].sum()

total_survived = total_first + total_second + total_third

print("The total passengers that survived in the ship: " + str(total_survived))
print( )
print("The number of passengers that survived who were travelling in first class are: " + str(training[training['Pclass']==1]['Survived'].sum()) + " out of a total: " + str(first))
print("The number of passengers that survived who were travelling in second class are: " + str(training[training['Pclass']==2]['Survived'].sum()) + " out of a total: " + str(second))
print("The number of passengers that survived who were travelling in third class are: " + str(training[training['Pclass']==3]['Survived'].sum()) + " out of a total: " + str(third))
```

```python
# Proportion of the those survived according to the class
proportion_first = total_first / total_survived
proportion_second = total_second / total_survived
proportion_third = total_third / total_survived

print("The proportion of first class that survived: " + str(proportion_first))
print("The proportion of second class that survived: " + str(proportion_second))
print("The proportion of third class that survived: " + str(proportion_third))
```
